## 1. Initialize the project

Fetch the working context: import the project 'tika'. Project is a placeholder for the code, data, and management of the data operations inside the digitalhub platform.

```python
import digitalhub as dh
PROJECT_NAME = "tika"
proj = dh.get_or_create_project(PROJECT_NAME)
```

## 2. Execution

### Deploy a text extraction service
Fetch the "tika" operation in the project. This function will use Apache Tika, a tool for extracting text from a variety of formats.

```python
tika_function = proj.get_function("tika") 
```

Then, execute the function. For more details about how to create, run, and view jobs inside to the digital hub platform see the <a href ="https://scc-digitalhub.github.io/docs/0.14/runtimes/python/">documentation </a>

```python
tika_run = tika_function.run("serve", service_ports = [{"port": 9998, "target_port": 9998}], wait=True)
```

As a result the 'tike' extraction service get deployed.

```python
service = tika_run.refresh().status.service
print("Service status:", service)
```

 Now fetch the URL of the service for API invocation.

```python
TIKA_URL = tika_run.status.to_dict()["service"]["url"]
print(TIKA_URL)
```

### Extract the text
We will not fetch a python function 'extract' from the project which will read an artifact from the platform's repository and leverage the Tika service to extract the textual content and write it to a HTML file.


```python
extract_function = proj.get_function("extract")
```
