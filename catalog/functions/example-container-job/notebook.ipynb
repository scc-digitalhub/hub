## 1. Initialize the project

Create the working context: data download project if not created already. Project is a placeholder for the code, data, and management of the data operations.

```python
import digitalhub as dh
PROJECT_NAME = "<YOUR_PROJECT_NAME>"
proj = dh.get_or_create_project(PROJECT_NAME)
```

Note: Make sure to replace <YOUR_PROJECT_NAME> with the actual name of your project before running the code.

## 3. Execution

Fetch the "example-container-job" operation in the project. 

```python
function = proj.get_function("example-container-job") 
```

Alternatively, one can create new function

```python
function = proj.new_function(
    kind="container",
    name="my_container_function_job",
    image="hello-world:latest",
)
```

We can now run the function and see the results. To do this we use the run method of the function.

```python
run = function.run(
    action="job",
    run_as_user="8877",
    wait=True,
)
```

Note: Wait for job to finish. Alternatively, using the Core Management UI, one can navigate to 'Runs' menu , select the corresponding 'run' instance and inspect the logs using 'Logs' tab.

The logs of platform execution can be viewed running the cell below.

```python
import base64
logs = base64.b64decode(run.logs()[0]['content']).decode()
print(logs)
```

## 3. Build Custom Container Function

The runtime enables users to build container-based functions with custom instructions and dependencies. This allows you to tailor the execution environment to your specific requirements.

**Usage Steps:**
1. **Create a new container function:**  
    Use `proj.new_function(kind="container", ...)` to define your function, specifying the base image.

2. **Add custom build instructions:**  
    Provide additional build steps (e.g., installing packages, copying files) using the `build_instructions` parameter.

3. **Specify dependencies:**  
    List required dependencies (such as Python packages or system libraries) in your build instructions or via a requirements file.

4. **Run the function:**  
    Execute your custom container job using the `run` method.

See the example below

```python
import digitalhub as dh
PROJECT_NAME = "<YOUR_PROJECT_NAME>"
proj = dh.get_or_create_project(PROJECT_NAME)


function = project.new_function(
    kind="container",
    name="my_function",
    base_image="python:3.11-slim",
)
run = function.run(
    action="build",
    instructions=["RUN apt-get update && apt-get install -y git"],
    wait=True,
)
```
