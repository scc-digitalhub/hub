## 1. Initialize the project

Create the working context: data download project if not created already. Project is a placeholder for the code, data, and management of the data operations.

```python
import digitalhub as dh
PROJECT_NAME = "<YOUR_PROJECT_NAME>"
proj = dh.get_or_create_project(PROJECT_NAME)
```

Note: Make sure to replace <YOUR_PROJECT_NAME> with the actual name of your project before running the code.

## 2. Execution

Fetch the "hello-world" operation in the project. 

```python
function_hello = proj.get_function(name="hello-world",
                            kind="python",
                            python_version="PYTHON3_10") 
```

Then, execute the function (locally) as specified with (local_execution=True) flag. This will execute the function as a single job in the local context. Without the local_execution flag the job will get executed on the platform. For more details about how to create, run, and view jobs inside to the digital hub platform see the <a href ="https://scc-digitalhub.github.io/docs/0.14/runtimes/python/">documentation </a> 

```python
function_hello.run(action='job', wait=True, local_execution=True)
```

As a result the 'hello world' message is logged.

Now, we can run the same function as a job in the platform by removing the 'local_execution' flag.


```python
run = function_hello.run(action='job', wait=True)
```

The logs of platform execution can be viewed running the cell below.

```python
import base64
logs = base64.b64decode(function_run.logs()[1]['content']).decode()
print(logs)
```

Note: Alternatively, using the Core Management UI, one can navigate to 'Runs' menu , select the corresponding 'run' instance and inspect the logs using 'Logs' tab.
